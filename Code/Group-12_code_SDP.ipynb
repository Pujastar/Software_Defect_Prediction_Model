We made  a model for Software Detefct Prection using machine learning algorithm.
In this project we have used four algorithm (ANN , PCA , KPCA , LDA) on five different datasets (cm1 , jm1 , kc1 , kc2 , pc1) and compared their accuracy ,  F1 score and ROC Curve Area.

```python
# Importing the libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score as roc
from sklearn.preprocessing import label_binarize
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.decomposition import PCA
from sklearn.decomposition import KernelPCA
```

Importing all the datasets .

```python
# Importing the dataset
datasetcm1 = pd.read_csv('cm1.csv')
datasetjm1 = pd.read_csv('jm1.csv')
datasetkc1 = pd.read_csv('kc1.csv')
datasetkc2 = pd.read_csv('kc2.csv')
datasetpc1 = pd.read_csv('pc1.csv')
datasets=[datasetcm1,datasetjm1,datasetkc1,datasetkc2,datasetpc1]
name=["cm1","jm1","kc1","kc2","pc1"]
dim_reds=["LDA","PCA","KPCA"]
```

We applied LDA , PCA  and KPCA algorihtm on 5 Dataset .
Then we compared the accuracy , precision , score and ROC curce area of those four algorihtm on five datasets (cm1,jm1,kc1,kc2,pc1) .


```python
for j in range(0,3):
  for i in range(0,5):
    print("Running for DATASET: ",name[i])
    print("Running for Method: ", dim_reds[j])
    X = datasets[i].iloc[:,:-1].values
    y = datasets[i].iloc[:, 21].values
    print("len of X",len(X))
    print("no of column",len(datasets[i].columns))
    # Encoding the Dependent Variable
    labelencoder_y = LabelEncoder()
    y = labelencoder_y.fit_transform(y)
    # Splitting the dataset into the Training set and Test set
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
    # Feature Scaling
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    # Applying Dimensionality reduction
    if dim_reds[j]=="LDA":
      lda = LDA(n_components = 6)
      X_train = lda.fit_transform(X_train, y_train)
      X_test = lda.transform(X_test)
    elif dim_reds[j]=="PCA":
      pca = PCA(n_components = 6)
      X_train = pca.fit_transform(X_train)
      X_test = pca.transform(X_test)
      explained_variance = pca.explained_variance_ratio_
    else:
      kpca = KernelPCA(n_components = 6, kernel = 'rbf')
      X_train = kpca.fit_transform(X_train)
      X_test = kpca.transform(X_test)
    # Fitting Logistic Regression to the Training set
    classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    classifier.fit(X_train, y_train)
    # Predicting the Test set results
    y_pred = classifier.predict(X_test)
    # Making the Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    print("confusionMatrix \n",cm)
    print("accuracy: ", (cm[0][0]+cm[1][1])/(cm[1][0]+cm[0][1]+cm[0][0]+cm[1][1]))
    precision=cm[0][0]/(cm[0][0]+cm[1][0])
    recall=cm[0][0]/(cm[0][0]+cm[0][1])
    print("precision: ",precision)
    print("recall: ",recall)
    print(" score: ",(2*recall*precision)/(recall+precision))
    # Finding Area Under ROC curve
    a = roc(y_test, y_pred, average='micro')
    #Processing data for plotting graph
    y = label_binarize(y, classes=[0, 1])
    n_classes = y.shape[1]
    onehotencoder = OneHotEncoder()
    y = onehotencoder.fit_transform(y).toarray()

    y_test = label_binarize(y_test, classes=[0, 1])
    onehotencoder = OneHotEncoder()
    y_test = onehotencoder.fit_transform(y_test).toarray()

    y_pred = label_binarize(y_pred, classes=[0, 1])
    onehotencoder = OneHotEncoder()
    y_pred = onehotencoder.fit_transform(y_pred).toarray()
    # Compute ROC curve and ROC area for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_pred[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    # Compute micro-average ROC curve and ROC area
    fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), y_pred.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

    # Plot of a ROC curve for a specific class
    plt.figure()
    lw = 2
    plt.plot(fpr[0], tpr[0], color='red',
            lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[0])
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic')
    plt.legend(loc="lower right")
    plt.show()
    print("\n")
```

We have applied Artifical Neural Network (ANN) algorithm on five dataset (cm1 , jm1 , kc1 , kc2 , pc1) and calculated the accuracy , precision ,score and ROC curve Area .

```python

import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import confusion_matrix
# Importing the dataset
for i in range(0,5):
  print("Running for DATASET: ",name[i])
  print("Running for Method: ANN")
  X = datasets[i].iloc[:,:-1].values
  y = datasets[i].iloc[:, 21].values

  # Encoding the Dependent Variable

  labelencoder_y = LabelEncoder()
  y = labelencoder_y.fit_transform(y)


  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)

  # Feature Scaling

  sc = StandardScaler()
  X_train = sc.fit_transform(X_train)
  X_test = sc.transform(X_test)

  #ANN model
  classifier = Sequential()
  classifier.add(Dense( 6, activation = 'relu', input_dim = 21))

  classifier.add(Dense( 6, activation = 'relu'))

  classifier.add(Dense(1, activation = 'sigmoid'))

  classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
  #fitting  the  model
  classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)

  y_pred = classifier.predict(X_test)
  y_pred = (y_pred > 0.3)
  cm = confusion_matrix(y_test, y_pred)
  # print(cm)
  print("confusionMatrix \n",cm)
  print("accuracy: ", (cm[0][0]+cm[1][1])/(cm[1][0]+cm[0][1]+cm[0][0]+cm[1][1]))
  precision=cm[0][0]/(cm[0][0]+cm[1][0])
  recall=cm[0][0]/(cm[0][0]+cm[0][1])
  print("precision: ",precision)
  print("recall: ",recall)
  print(" score: ",(2*recall*precision)/(recall+precision))
  # Finding Area Under ROC curve
  a = roc(y_test, y_pred, average='micro')
  #Processing data for plotting graph
  y = label_binarize(y, classes=[0, 1])
  n_classes = y.shape[1]
  onehotencoder = OneHotEncoder()
  y = onehotencoder.fit_transform(y).toarray()

  y_test = label_binarize(y_test, classes=[0, 1])
  onehotencoder = OneHotEncoder()
  y_test = onehotencoder.fit_transform(y_test).toarray()

  y_pred = label_binarize(y_pred, classes=[0, 1])
  onehotencoder = OneHotEncoder()
  y_pred = onehotencoder.fit_transform(y_pred).toarray()
  # Compute ROC curve and ROC area for each class
  fpr = dict()
  tpr = dict()
  roc_auc = dict()
  for i in range(n_classes):
      fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_pred[:, i])
      roc_auc[i] = auc(fpr[i], tpr[i])
  # Compute micro-average ROC curve and ROC area
  fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), y_pred.ravel())
  roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

  # Plot of a ROC curve for a specific class
  plt.figure()
  lw = 2
  plt.plot(fpr[0], tpr[0], color='red',
          lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[0])
  plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
  plt.xlim([0.0, 1.0])
  plt.ylim([0.0, 1.05])
  plt.xlabel('False Positive Rate')
  plt.ylabel('True Positive Rate')
  plt.title('Receiver operating characteristic')
  plt.legend(loc="lower right")
  plt.show()
  print("\n")
```
